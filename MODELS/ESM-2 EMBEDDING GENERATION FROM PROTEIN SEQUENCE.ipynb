```python
import requests
from time import sleep

uniprot_ids =NEGATIVE['Accession (UniProtKB)'].tolist()

# Base URL for UniProt API
base_url = "https://rest.uniprot.org/uniprotkb"

# File to save sequences
output_file = "unique_NEGATIVE_NGLYCO_PROTEIN__.fasta"
error_file = "uniqueNEGATIVE_NGLYCO_PROTEIN__ERROR.txt"

# Batch size for processing
batch_size = 500  # Adjust based on server and network limits
total_ids = len(uniprot_ids)
print(f"Total UniProt IDs (including duplicates): {total_ids}")

# Process IDs in batches
with open(output_file, "w") as fasta_file, open(error_file, "w") as error_log:
    for i in range(0, total_ids, batch_size):
        batch = uniprot_ids[i:i + batch_size]
        print(f"Processing batch {i // batch_size + 1} of {total_ids // batch_size + 1}")
        
        for uniprot_id in batch:
            url = f"{base_url}/{uniprot_id}.fasta"
            try:
                response = requests.get(url)
                if response.status_code == 200:
                    # Write sequence to the FASTA file
                    fasta_file.write(response.text)
                    print(f"Sequence for {uniprot_id} retrieved.")
                else:
                    # Log ID in case of failure
                    error_log.write(f"{uniprot_id}\n")
                    print(f"Failed to retrieve sequence for {uniprot_id}: {response.status_code}")
            except Exception as e:
                # Log any unexpected errors
                error_log.write(f"{uniprot_id} - Error: {str(e)}\n")
                print(f"Error retrieving {uniprot_id}: {str(e)}")
        
        # Pause between batches to avoid overwhelming the server
        sleep(5)

print(f"All sequences have been processed. Results saved in '{output_file}'. Errors logged in '{error_file}'.")
```

```python
#import pandas as pd
from Bio import SeqIO

def extract_window(sequence, position, window_size=25):
    """
    Extracts a window of amino acids centered around the given position.
    
    Parameters:
    sequence (str): Protein sequence.
    position (int): 1-based index of the central amino acid.
    window_size (int): Number of residues before and after the central position.
    
    Returns:
    str: Extracted window sequence.
    """
    start = max(0, position - window_size - 1)  # Convert to 0-based index
    end = min(len(sequence), position + window_size)
    return sequence[start:end]

# Read protein sequences from FASTA file
#def load_fasta(fasta_file):
    sequences = {}
    for record in SeqIO.parse(fasta_file, "fasta"):
        # Extract the PID (Protein ID) from the FASTA header (e.g., >sp|A0A075B759|PAL4E_HUMAN)
        pid = record.id.split("|")[1]  # PID is between the first and second '|'
        sequences[pid] = str(record.seq)
    return sequences

# Read positions from Excel file and extract windows
def process_positions(fasta_file, excel_file, output_file):
    sequences = load_fasta(fasta_file)
    df = pd.read_excel(excel_file)
    
    extracted_data = []
    for _, row in df.iterrows():
        label = row['label']
        protein_id = row['PID']
        position = row['position']
        
        
        if protein_id in sequences:
            window = extract_window(sequences[protein_id], position, window_size=25)  # Use 25 residues
            extracted_data.append([protein_id, position, window, label])
    
    # Create DataFrame and save to CSV
    result_df = pd.DataFrame(extracted_data, columns=['label','protein_id', 'position', 'sequence_window' ])
    result_df.to_csv(output_file, index=False)

# Example usage:
fasta_file = "whole_unique_NEGATIVE_NGLYCO_PROTEIN__.fasta"
excel_file = "LOCATION_FOR_NEGATIVE_Proteins.xlsx"
output_file = "LOCATION_FOR_NEGATIVE_Proteins.csv"
process_positions(fasta_file, excel_file, output_file)
```

# splitting single fasta into multiple fasta

```python
from Bio import SeqIO
import os

# Specify the input FASTA file path
input_file =r"C:\Users\harsh\Downloads\final_data_for_glycosylation\neg\unique_NEGATIVE_NGLYCO_PROTEIN__.fasta"
# Specify the output folder to save the smaller FASTA files
output_folder =r"C:\Users\harsh\Downloads\final_data_for_glycosylation\neg"
# Make sure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Read the sequences from the input FASTA file
records = list(SeqIO.parse(input_file, "fasta"))

# Set the chunk size (25 sequences per file)
chunk_size = 25

# Split the records into chunks of 25 sequences
for i in range(0, len(records), chunk_size):
    # Create a subset of the records for the current chunk
    chunk = records[i:i + chunk_size]
    
    # Define the output file for the current chunk (renamed as file1.fasta, file2.fasta, etc.)
    output_file = os.path.join(output_folder, f"file{i//chunk_size + 1}.fasta")
    
    # Write the chunk to the output file
    with open(output_file, "w") as output:
        SeqIO.write(chunk, output, "fasta")

print("FASTA file has been split into files with 25 sequences each and renamed successfully.")

```

```python
import torch
from transformers import AutoTokenizer, AutoModel
from Bio import SeqIO
import os
import pandas as pd

# Set directory paths

input_dir = r"C:\Users\harsh\Downloads\final_data_for_glycosylation\unique_negative"  # Replace with actual path
output_dir = r"C:\Users\harsh\Downloads\final_data_for_glycosylation\unique_negative"  # Replace with actual path

# Change working directory to input directory
os.chdir(input_dir)

# Load the ESM-2 tokenizer and model
print("Loading the tokenizer and model...")
model_name = "facebook/esm2_t36_3B_UR50D"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
print("Tokenizer and model loaded successfully.")

# Input FASTA file
fasta_file = os.path.join(input_dir, "unique_NEGATIVE_NGLYCO_PROTEIN__.fasta")  # Replace "file1.fasta" with your actual file name


# Process each sequence in the FASTA file
for seq_record in SeqIO.parse(fasta_file, "fasta"):
    protein_id = seq_record.id.split("|")[1]
    sequence = str(seq_record.seq).replace("U", "X").replace("Z", "X").replace("O", "X").replace("B", "X")
    sequence_spaced = ' '.join(sequence)
    
    
    # Tokenize and compute embeddings
    inputs = tokenizer(sequence_spaced, return_tensors="pt", padding=True, truncation=True)

    # Pass the tokenized input through the model to get the hidden states
    with torch.no_grad():
        outputs = model(**inputs)
        hidden_states = outputs.last_hidden_state

    # Extract valid embeddings and save to CSV
    valid_embeddings = hidden_states[0][:len(sequence)].numpy()
    output_file = os.path.join(output_dir, f"{protein_id}_ESM_.csv")

    # Writing the features to CSV with the correct headers
    
    with open(output_file, "w") as fp:
        headers = ["Amino_Acid", "Position"] + [f"Feature_{i}" for i in range(1, valid_embeddings.shape[1] + 1)]
        fp.write(",".join(headers) + "\n")
        
        for i, embedding in enumerate(valid_embeddings):
            feature_vector = ','.join(map(str, embedding.tolist()))
            fp.write(f"{sequence[i]},{i+1},{feature_vector}\n")  # Add position as i+1

    print(f"Finished processing protein: {protein_id}")

print("All proteins processed successfully.")
```

# FEATURE EXTRACTION FROM  GENERATED EMBEDDING FILES 

```python
import pandas as pd
import os

def update_position_pid_file(position_pid_file, directory, output_file):
    """
    Updates the position PID file by adding features from corresponding PID CSV files.

    Args:
        position_pid_file (str): Path to the position PID file (input CSV with 'PID' and 'Position').
        directory (str): Directory containing PID CSV files.
        output_file (str): Path to save the updated position PID file (output CSV).
    """
    # Read the position PID file
    position_pid_df = pd.read_csv(position_pid_file)
    print(f"Loaded position PID file: {position_pid_file}, {position_pid_df.shape[0]} rows")

    # Create a dictionary to store the data from PID.csv files (cache)
    pid_data_cache = {}

    # List to store rows with extracted features
    rows_to_add = []

    # Iterate through each row in the position PID file
    for index, row in position_pid_df.iterrows():
        pid = row['PID']  # Get the PID
        position = row['Position']  # Get the Position
        print(f"Processing PID: {pid}, Position: {position}")

        # Check if the PID file is already loaded in the cache
        if pid not in pid_data_cache:
            # Construct the file name for the PID
            pid_file = os.path.join(directory, f"{pid}.csv")

            # Load the PID.csv file if it exists
            if os.path.exists(pid_file):
                print(f"Found file for {pid} at {pid_file}")
                pid_data_cache[pid] = pd.read_csv(pid_file)
            else:
                print(f"File not found for PID: {pid}")
                continue

        # Get the corresponding PID DataFrame
        pid_df = pid_data_cache[pid]

        # Filter the row corresponding to the given position
        extracted_row = pid_df[pid_df['Position'] == position]

        if not extracted_row.empty:
            # Drop 'PID' and 'Position' columns from the extracted row (already in position_pid_df)
            features_to_add = extracted_row.drop(columns=['PID', 'Position'], errors='ignore')

            # Add the new features as a dictionary to the list
            new_row = features_to_add.iloc[0].to_dict()  # Convert row to dictionary
            new_row['index'] = index  # Keep track of the row index
            rows_to_add.append(new_row)
            print(f"Prepared row for PID {pid} at Position {position} for addition.")
        else:
            print(f"Position {position} not found in the PID file for {pid}.")

    # If rows to add are collected, process them
    if rows_to_add:
        print(f"Rows to add: {len(rows_to_add)}")
        # Convert the list of rows into a DataFrame
        new_features_df = pd.DataFrame(rows_to_add)

        # Merge the new features into the original DataFrame
        for column in new_features_df.columns:
            if column != 'index':  # Skip the 'index' column
                # Add the column if it doesn't exist
                if column not in position_pid_df.columns:
                    position_pid_df[column] = None
                # Assign values based on the index
                position_pid_df.loc[new_features_df['index'], column] = new_features_df[column].values

    # Write the updated position PID file back to disk
    position_pid_df.to_csv(output_file, index=False)
    print(f"Output file saved to: {output_file}")

# Example usage:
position_pid_file ="C:\Users\harsh\Downloads\final_data_for_glycosylation\LOCATION_FOR_NEGATIVE_Proteins.csv"  # Path to the position PID file
directory =r"C:\Users\harsh\Downloads\final_data_for_glycosylation\unique_negative" # Directory containing PID CSV files
output_file =r"C:\Users\harsh\Downloads\final_data_for_glycosylation\final_unique_negative_glycoproteins_features.xlsx"  # Path to save the updated file


# Run the function
update_position_pid_file(position_pid_file, directory, output_file)
```
